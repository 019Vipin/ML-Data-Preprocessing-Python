Importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the dataset to a Pandas DataFrame
import pandas as pd
dataset = pd.read_csv('/content/Placement_Dataset.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

Central Tendencies:

1. Mean
2. Median
3. Mode

```python
# analyse the distribution of data in the salary
import matplotlib.pyplot as plt
import seaborn as sns
fig, ax = plt.subplots(figsize=(10,5))
sns.histplot(dataset.salary)
```


Replace the missing values with Median value

```python
dataset['salary'] = dataset['salary'].fillna(dataset['salary'].median())
```

```python
dataset.isnull().sum()
```

```python
# filling missing values with Mean value:
dataset['salary'] = dataset['salary'].fillna(dataset['salary'].mean())
```

```python
# filling missing values with Mean value:
dataset['salary'] = dataset['salary'].fillna(dataset['salary'].mode())
```
